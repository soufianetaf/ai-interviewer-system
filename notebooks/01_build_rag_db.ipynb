```python
import sys
import os
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# 1. Importer la configuration
sys.path.append(os.path.abspath('..'))
from config.settings import FAISS_INDEX_PATH, EMBEDDING_MODEL_NAME

print(f" Initialisation du modèle d'embedding : {EMBEDDING_MODEL_NAME}")

# 2. Initialiser le modèle d'embedding via LangChain
embeddings = HuggingFaceEmbeddings(model_name=EMBEDDING_MODEL_NAME)

# 3. Préparation de la documentation officielle (Les données brutes)
documents_rh = [
    "Une clé primaire est une contrainte de base de données qui identifie de manière unique chaque ligne d'une table. Elle ne peut pas être NULL.",
    "Un index en SQL améliore la vitesse des opérations de recherche sur une table, mais ralentit les opérations d'insertion et de mise à jour.",
    "Une jointure LEFT JOIN retourne toutes les lignes de la table de gauche, et les correspondances de la table de droite. S'il n'y a pas de correspondance, des valeurs NULL sont renvoyées.",
    "Une requête lente peut être optimisée en ajoutant des index sur les colonnes filtrées (WHERE) ou jointes (JOIN), et en évitant le SELECT *."
]

print(" Création de la base vectorielle FAISS en mémoire...")
# 4. Création de la base vectorielle (tout se fait en RAM d'abord)
vectorstore = FAISS.from_texts(texts=documents_rh, embedding=embeddings)

print(f" Sauvegarde de la base dans Unity Catalog : {FAISS_INDEX_PATH}")
# 5. Sauvegarde sur le disque (crée un fichier .faiss et un .pkl sans faire planter le serveur)
vectorstore.save_local(FAISS_INDEX_PATH)

print(" Succès ! La base de connaissances est prête.")
```
