```python
from langchain_core.documents import Document
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from config.settings import FAISS_INDEX_PATH, EMBEDDING_MODEL_NAME

print(" Préparation des nouvelles connaissances RH...")

# 1. Notre nouvelle base de connaissances complète
nouvelles_connaissances = [
    Document(page_content="[Clé Primaire] Une clé primaire est une contrainte qui identifie de manière unique chaque enregistrement d'une table de base de données. Elle doit contenir des valeurs uniques et ne peut pas contenir de valeurs NULL."),
    Document(page_content="[Index SQL] Un index est une structure de données utilisée pour accélérer considérablement la récupération de données (les SELECT) dans une table. Il ne ralentit pas la base, bien qu'il puisse légèrement ralentir les opérations d'écriture (INSERT/UPDATE) car l'index doit être mis à jour."),
    Document(page_content="[Optimisation SQL] Pour optimiser une requête SQL lente, les meilleures pratiques sont : 1) Créer des index pertinents sur les colonnes filtrées (WHERE, JOIN). 2) Éviter l'utilisation de 'SELECT *' et ne spécifier que les colonnes nécessaires. 3) Analyser le plan d'exécution (EXPLAIN).")
]

# 2. Chargement du modèle d'embedding
embeddings = HuggingFaceEmbeddings(model_name=EMBEDDING_MODEL_NAME)

# 3. Création et écrasement de l'ancienne base
vectorstore = FAISS.from_documents(nouvelles_connaissances, embeddings)
vectorstore.save_local(FAISS_INDEX_PATH)

print(" Base FAISS mise à jour avec succès dans Unity Catalog !")
```
